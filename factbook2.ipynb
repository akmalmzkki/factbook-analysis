```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

# Preprocessing data

```python
# Menghilangkan spasi pada awal nama kolom
df.columns = df.columns.str.strip()
df.info()
```

```python
columns1 = ['GDP', 'GDP per capita', 'Imports', 'Exports', 'Electricity production', 'Internet users']
for column in columns1:
    df[column] = df[column].str.replace(',', '').str.replace('.', '').str.replace('$', '')
    df[column] = df[column].str[:-3]
    df[column] = df[column].astype(np.int64)
df
```

```python
columns2 = ['Area', 'Electricity consumption', 'Oil consumption', 'Population', 'Labor force', 'Highways']
for column in columns2:
    df[column] = df[column].str.replace(' ', '0')
    df[column] = df[column].str.replace(',', '').astype(np.int64)
df.info()
```

```python
df['Current account balance'] = df['Current account balance'].str.replace('$', '').str.replace(',', '').str.replace('(', '').str.replace(')', '').str.replace('.', '')
df['Current account balance'] = df['Current account balance'].astype(np.int64)
df['Current account balance'] = np.int64(df['Current account balance'] / 100)
```

```python
df.info()
```

```python
columns3 = ['Industrial production growth rate', 'Investment', 'Military expenditures', 'Public debt', 'Unemployment rate']
for column in columns3:
    df[column] = df[column].str.replace(' ', '0').astype(np.float64)
df.info()
```

```python
df['Railways'] = df['Railways'].str.replace(' ', '0').astype(np.int64)
df.info()
```

```python
df['Reserves of foreign exchange & gold'] = df['Reserves of foreign exchange & gold'].str.replace(' ', '0').str.replace(',', '').str.replace('.', '')
df['Reserves of foreign exchange & gold'] = df['Reserves of foreign exchange & gold'].astype(np.int64)
df['Reserves of foreign exchange & gold'] = np.int64(df['Reserves of foreign exchange & gold'] / 100)
```

```python
df.info()
```

```python
df['Natural gas consumption'] = df['Natural gas consumption'].str.replace(' ', '0').str.replace('-', '0').str.replace('.', '').str.replace(',', '')
df['Natural gas consumption'] = df['Natural gas consumption'].astype(np.int64)
df['Natural gas consumption'] = np.int64(df['Natural gas consumption'] / 100)
```
